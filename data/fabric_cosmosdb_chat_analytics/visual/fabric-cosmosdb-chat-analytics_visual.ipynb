```python
# Load the "chat_analysis_results" table into a dataframe
df = spark.read.table("fabric_cosmosdb_chat_lakehouse.chat_analysis_results")

# Display a pie chart for the data in the "chat_analysis_results" table
import plotly.express as px

fig = px.pie(df, names='sum_liked_price', title='Pie Chart of Sum of Liked Price', template='plotly_dark')
fig.show()
```

```python
%%chat
# create a heatmap to display liked and disliked distibutions in chat_analysis_results.
```

```python
import pandas as pd
import plotly.express as px

p_df = df.toPandas()
# Selecting only relevant columns for heatmap
heatmap_data = p_df[['sum_liked_price', 'sum_liked_size', 'sum_liked_usability', 'sum_liked_color',
                   'sum_liked_brand', 'sum_liked_design', 'sum_disliked_price', 'sum_disliked_size',
                   'sum_disliked_usability', 'sum_disliked_color', 'sum_disliked_brand', 
                   'sum_disliked_design']]

fig = px.imshow(heatmap_data.corr(), x=heatmap_data.columns, y=heatmap_data.columns, 
                template='plotly_dark', title='Liked and Disliked Distributions Heatmap')
fig.show()
```
